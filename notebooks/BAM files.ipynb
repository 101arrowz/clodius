```python
%%time
import numpy as np
n_reads = 10000
values = {}
for reference in samfile.references:
    values[reference] = np.zeros(samfile.get_reference_length(reference))
    
for i,read in enumerate(samfile):
    if i >= n_reads:
        break
    query_seq = read.query_sequence

    differences = []
    try:
        for counter, (qpos, rpos, ref_base) in enumerate(read.get_aligned_pairs(with_seq=True)):
            # print('qpos', qpos)
            # inferred from the pysam source code:
            # https://github.com/pysam-developers/pysam/blob/3defba98911d99abf8c14a483e979431f069a9d2/pysam/libcalignedsegment.pyx
            # and GitHub issue:
            # https://github.com/pysam-developers/pysam/issues/163
            #print('qpos, rpos, ref_base', qpos, rpos, ref_base)
            values[i][0] += 1
            if rpos is None:
                values[i][1] += 1
            elif qpos is None:
                values[i][2] += 1
            elif ref_base.islower():
                if query_seq[qpos] == 'A':
                    values[i][3] += 1
                if query_seq[qpos] == 'C':
                    values[i][4] += 1
                if query_seq[qpos] == 'G':
                    values[i][5] += 1
                if query_seq[qpos] == 'T':
                    values[i][6] += 1
    except ValueError as ve:
        # probably lacked an MD string
        pass
# print(i)
#    print(read)
```
